```python
#insertion and deletion

arr = [1,2,3,4,5]

#insert
pos = int(input("Enter the position you want to insert the value (0 based index)"))
element = int(input("Enter the element you want to insert"))

if pos >= 0 and pos <= len(arr):
    arr.insert(pos,element)
    print("After Insertion",arr)
else:
    print("Invalid position for insertion",arr)

#deletion
del_pos = int(input("Enter the position you want to delete the value (0 based index)"))   

if pos >= 0 and pos <= len(arr):
    arr.pop(del_pos)
    print("updated",arr)
else:
    print("Invalid position for deletion",arr)
```
